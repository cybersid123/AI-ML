#Series- Data Type

```python
import numpy as np
import pandas as pd
```

#Create a Series

```python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d = {'a':10, 'b':20, 'c':30}
```

```python
pd.Series(data = my_list)
```

```python
type(pd.Series(data = my_list))
```

```python
pd.Series(data = my_list, index = labels)
```

```python
#using array
```

```python
pd.Series(arr)
```

```python
pd.Series(arr, index = labels)
```

```python
#using dict
```

```python
pd.Series(data = d)
```
