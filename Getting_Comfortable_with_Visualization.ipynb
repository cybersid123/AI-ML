#Seaborn

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('heart.csv')

df.head()

```

#Distribution Plot

```python
sns.distplot(df['age'])

sns.displot(df['age'])

sns.displot(df['age'], bins = 40, color='red')

sns.countplot(data = df, x='target')

sns.countplot(x='cp', data =df, hue='sex', palatte= 'red')

sns.boxplot(x='sex', y='age',data=df)


```

#Scatter Plot

```python
sns.scatterplot(data = df, x='chol', y='trestbps', hue='sex')

iris = pd.read_csv('iris.csv')
iris.head()

sns.pairplot(iris, hue='species')
```
